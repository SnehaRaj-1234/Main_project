```python
import shutil
import os

src_img_dir = "/kaggle/input/temp-data-facerec/PImg/PImg/Familiar"
src_lbl_dir = "/kaggle/input/temp-data-facerec/Labels/Labels"
dst_img_dir = "/kaggle/working/dataset/images"
dst_lbl_dir = "/kaggle/working/dataset/labels"

# Create writable directories
os.makedirs(dst_img_dir, exist_ok=True)
os.makedirs(dst_lbl_dir, exist_ok=True)

# Copy images
for file in os.listdir(src_img_dir):
    shutil.copy(os.path.join(src_img_dir, file), dst_img_dir)

# Copy labels (excluding 'classes.txt')
for file in os.listdir(src_lbl_dir):
    if file != "classes.txt":  
        shutil.copy(os.path.join(src_lbl_dir, file), dst_lbl_dir)

print("✅ Dataset copied without 'classes.txt'. Use '/kaggle/input/video-img/PImg' for training.")
```

```python
from ultralytics import YOLO

yaml_path = "/kaggle/working/data.yaml"

# Load YOLOv5s with COCO weights and train on your dataset
model = YOLO("yolov5s.pt")  # COCO-pretrained YOLOv5s
model.train(data=yaml_path, epochs=50, batch=16, imgsz=640, name="custom_familiar_detector", pretrained=True)

```

```python
# Evaluate the trained YOLOv5 model
metrics = model.val()
```
